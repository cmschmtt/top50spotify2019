```python
import numpy
import matplotlib.pyplot as plt
import csv
from itertools import groupby
```

```python
data = []
with open('top50.csv', mode='r', encoding='ISO-8859-1') as file:
    csv_reader = csv.reader(file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        data_sample = {}
        if line_count==0:
            keys = row[1:]
            keys.insert(0,'ID')
            line_count += 1
        else:
            for i in range(len(keys)):
                data_sample[keys[i]] = row[i]
            data.append(data_sample)
            line_count += 1
```

```python
name = []
artist = []
genre = []
bpm = []
energy = []
dance = []
loudness = []
liveness = []
valence = []
length = []
acoustic = []
speech = []
popularity = []
for i in range(len(data)):
    name.append(data[i]['Track.Name'])
    artist.append(data[i]['Artist.Name'])
    genre.append(data[i]['Energy'])
    bpm.append(int(data[i]['Beats.Per.Minute']))
    energy.append(int(data[i]['Energy']))
    dance.append(int(data[i]['Danceability']))
    loudness.append(int(data[i]['Loudness..dB..']))
    liveness.append(int(data[i]['Liveness']))
    valence.append(int(data[i]['Valence.']))
    length.append(int(data[i]['Length.']))
    acoustic.append(int(data[i]['Acousticness..']))
    speech.append(int(data[i]['Speechiness.']))
    popularity.append(int(data[i]['Popularity']))
```

```python
b = [len(list(group)) for key, group in groupby()]
```

```python
fig = plt.figure(figsize=(16,12))

# BPM
ax = fig.add_subplot(241)
ax.hist(bpm, bins=20)
plt.xlabel('BPM')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of BPM')

# Length
ax = fig.add_subplot(242)
ax.hist(length, bins=20)
plt.xlabel('Length [seconds]')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Length')

# Danceability
ax = fig.add_subplot(243)
ax.hist(dance, bins=20)
plt.xlabel('Danceability')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Danceability')

# Loudness
ax = fig.add_subplot(244)
ax.hist(loudness, bins=20)
plt.xlabel('Loudness [dB]')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Loudness')

# Liveness
ax = fig.add_subplot(245)
ax.hist(liveness, bins=20)
plt.xlabel('Liveness (% likelihood track is live recording)')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Liveness')

# Valence
ax = fig.add_subplot(246)
ax.hist(valence, bins=20)
plt.xlabel('Valence (positive mood)')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Valence')

# Acousticness
ax = fig.add_subplot(247)
ax.hist(acoustic, bins=20)
plt.xlabel('Acousticness')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Acousticness')

# Speechiness
ax = fig.add_subplot(248)
ax.hist(speech, bins=20)
plt.xlabel('Speechness')
plt.ylabel('Number of tracks')
ax.title.set_text('Histogram of Speechness (rapiness)')
plt.savefig('histogram.png')
plt.show()
```
